```python
import pandas as pd
import numpy as np
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams.update(mpl.rcParamsDefault)
# mpl.rcParams["text.usetex"] = True
mpl.rcParams["font.family"] = "serif"
mpl.rcParams["font.size"] = "12"
```

```python
ts_prices_btc = pd.read_csv("Data/Time_Series/prices/bitcoin_19_22.csv",index_col='date',parse_dates=['date'])
```

```python
crypto = 'crypto_bitcoin'
model ='VADER'
df= pd.read_csv(f"Data/Time_Series/{crypto}/{model}_sts_19_22.csv",index_col='posted_on',parse_dates=['posted_on'])
```

```python
ts_prices_btc= ts_prices_btc.join(df)
```

```python
# plt.figure(figsize=(14, 8))
fig, ax1 = plt.subplots(figsize=(12, 6))
color = 'red'
ax1.set_xlabel('date')
ax1.set_ylabel('Activity Count', color='black')
ax1.plot(ts_prices_btc.index, ts_prices_btc[['total_posts']], color='green', label=f'Reddit Posts')
# ax1.plot(df.index, df[['total_comments']], color='yellow', label=f'Reddit Posts')
ax1.tick_params(axis='y', labelcolor='black')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'blue'
ax2.set_ylabel('Price (Log-Scale)', color='black')  # we already handled the x-label with ax1
ax2.plot(ts_prices_btc.index, np.log(ts_prices_btc.price), color='orange', label=f'BTC Closing Price (USD)')
ax2.tick_params(axis='y', labelcolor='black')
ax1.tick_params(axis='x', labelrotation=90)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.3), fancybox=True, shadow=True, ncol=3)
ax1.set_xlabel('Date', color='black',labelpad=20) 

plt.title(f'Sampled Reddit Activity and Closing BTC Price')
# plt.xticks(ts_pos_sub.index)
plt.grid()
plt.show()
```
